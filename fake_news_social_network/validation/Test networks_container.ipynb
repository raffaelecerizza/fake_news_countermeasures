```python
network = Network(n_nodes=185, 
            alpha=0.25, beta=0.5, 
            delta_in=1, delta_out=1, 
            homophily=0.8, n_commons=180, 
            n_influencers=5, n_bots=0, n_fact_checkers=0,
            prob_complaint=0.1, 
            prob_infection=0.5, 
            prob_vaccination=0.1,
            prob_cure=0.1, 
            prob_influencer=0.1, 
            exp_decay=True, user_block=False, 
            prob_echo=0.0)
network.create_graph()
initial_infected_nodes = random.sample(range(0, 184), 7)
network.set_initial_infected_nodes(initial_infected_nodes)
print(initial_infected_nodes)
```

```python
y_pred_list = []
network.prob_complaint = 0.1
network.prob_infection = 0.1
network.prob_vaccination = 0.1
network.prob_cure = 0.1
for i in range(network.n_nodes):
    network.nodes[i].prob_complaint = network.prob_complaint
    network.nodes[i].prob_infection = network.prob_infection 
    network.nodes[i].prob_vaccination = network.prob_vaccination 
    network.nodes[i].prob_cure = network.prob_cure 
for i in range(253):
    network.update_nodes()
    y_pred = network.perc_infected[i]
    y_pred_list.append(y_pred)
print(y_pred_list)
```
